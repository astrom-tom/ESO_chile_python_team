### Excercise 1
Create a figure that contains one axis showing
- a sine and a cosine function, with the x-axis ranging from -$\pi$ to $\pi$ and the y-axis from -1.05 to 1.05.
- The figure should have meaningful axis labels and a legend that does not overlap with the plot. 
- The two plotted lines have to be distinguishable not only by their colour.
- Labels and legend have to be rendered in TeX
- Save the figure to disk (Hint: [matplotlib documentation](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.savefig.html))

````python
# solution 
# 1) create x-data 
x = np.linspace(-np.pi, np.pi, 100)
# 2) create figure 
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, np.sin(x), ':', label='$y = \sin(x)$')
ax.plot(x, np.cos(x), '--', label='$y = \cos(x)$')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.legend()
ax.set_xlim(-np.pi, np.pi)
ax.set_ylim(-1.05, 1.05)
# fig.savefig('your/favourite/directory/figure.png')
````

There are two ways to use Matplotlib.

For the plot above, we have used the **object-oriented style**.
   - We explicitely created figures and axes, and then called methods on them.

We could also have used the so-called **pyplot-style**
   - Pyplot automatically creates and manages the figures and axes. 

I personally prefer the object-oriented style, as it is easier to keep control over complicated figures. Below is an example of the same figure in pyplot style. 

````python
plt.plot(xdata, xdata, label='y=x')
plt.plot(xdata, xdata**2, label='y=x^2')
plt.plot(xdata, xdata**3, label='y=x^3')
plt.xlabel('x [arbitrary units]')
plt.ylabel('y [arbitrary units]')
plt.title('A simple line plot')
plt.legend()
plt.grid('on')
plt.xlim(0, 4)
plt.ylim(0, 70)
````

### Error bars

You will not get around them ;) 

````python
# Again, we'll first "fabricate" some data
xdata = np.arange(10, 120, 27)
xerr = 5
ydata = xdata*0.2 + 42.
yerr_lo = np.random.randint(5, 15, len(ydata))
yerr_hi = np.random.randint(5, 15, len(ydata))

# and then plot
fig = plt.figure()
ax = fig.add_subplot(111)
ax.errorbar(xdata, ydata,
            xerr=xerr, yerr=[yerr_lo, yerr_hi],
            fmt=' ', capsize=5)
````

### Histograms

````python
# generating some random distributions
data1 = np.random.normal(50, 10, 200)
data2 = np.random.uniform(15, 70, 200)

# plot
fig = plt.figure()
ax1 = fig.add_subplot(111)
N, bins, patches = ax1.hist(data1, bins=25, label='normal')
N, bins, patches = ax1.hist(data2, bins=25, label='uniform')
ax1.legend()
ax1.set_ylabel('Counts')
ax1.set_xlabel('Data')
````

````python
# plot
fig = plt.figure()
ax1 = fig.add_subplot(111)
N, bins, patches = ax1.hist(data1, bins=25, label='normal', histtype='step') 
N, bins, patches = ax1.hist(data2, bins=25, label='uniform', histtype='step')
ax1.legend()
ax1.set_ylabel('Counts')
ax1.set_xlabel('Data')
````

````python
# custom bins 
bins1 = np.linspace(0, 85, 15)
bins2 = np.array([0, 20, 35, 55, 75, 90])

fig = plt.figure()
ax1 = fig.add_subplot(111)
N, bins, patches = ax1.hist(data1, bins=bins1, histtype='stepfilled') 
N, bins, patches = ax1.hist(data1, bins=bins2, histtype='step')
ax1.set_ylabel('Counts')
ax1.set_xlabel('Data')
````

### Python functionalities 

- Use python syntax to make plotting more efficient
- Loops
- Conditionals
- Lists
- ...

````python
xrange = np.linspace(-5, 5, 100)

fig = plt.figure()
ax1 = fig.add_subplot(111)
for i in range(8):
    ax1.plot(xrange, np.cos(xrange)+i, label='Number '+str(i+1))
ax1.legend()
````

````python
fig = plt.figure()
ax1 = fig.add_subplot(111)
for i in range(8):
    # colour code even-odd
    if i%2==0:
        ax1.plot(xrange, np.cos(xrange)+i, label='Number '+str(i+1), color='b')
    else:
        ax1.plot(xrange, np.cos(xrange)+i, label='Number '+str(i+1), color='r')
````

![flags](https://vignette.wikia.nocookie.net/bigbangtheory/images/b/b9/Funwithflags_zpsf57e3cf1.jpg/revision/latest?cb=20141216003140)

````python
bars = [1,2,3]
heights = [2,1,3]

with plt.xkcd():
    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    ax1.bar(bars, heights, width=0.5, color='k')
    ax1.set_xticks([1, 2, 3])
    ax1.set_ylabel('AMOUNT OF BLACK INK BY PANEL')
    ax1.set_yticks([])
````

Using the with-statement above, the style will only be applied temporary. If you use the statement without the `with`, you'll be stuck with the xkcd-style for the remainder of the notebook. 

#### A word about stylesheets
We can easily change back to the default [stylesheet](https://matplotlib.org/3.1.0/tutorials/introductory/customizing.html). 
Style sheets are amazing to have consistent plots. You can use them for a whole notebook (e.g. `plt.style.use('default')` or temporary with a `with` statement, as I did above. Once you're a bit more familiar with plotting, you can create your own style sheets. For example, I two distinct style sheets: one for publications in A&A, one for presenation with larger font sizes and labels. 

````python
plt.style.use('default')
bars = [1,2,3]
heights = [2,1,3]
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.bar(bars, heights, width=0.5, color='k')
ax1.set_xticks([1, 2, 3])
ax1.set_ylabel('AMOUNT OF BLACK INK BY PANEL')
ax1.set_yticks([])
````

### Excercise 2: Plot a Hertzsprung-Russell diagram
The goal of this excercise is to create a Hertzsprung-Russell diagram from tabular data. 

![hr](hr.png)

#### Prelim: read in the data
Data can be found [here](http://burro.astr.cwru.edu/Academics/Astr221/HW/HW5/yaletrigplx.dat). 
The columns are 
- column 1: star ID number
- column 2: apparent V magnitude
- column 3: observed B-V color
- column 4: observed parallax (in arcsec)
- column 5: uncertainty in parallax (in milliarcsec)

#### 2a) Plot the distribution of the observed parallaxes
- Determine a suitable bin size
- Make a histogram

#### 2b) The actual HR diagram
- Plot magnitude versus colour 
- As usual, in astronomy the magnitude axis is inverted
- Each point should be coloured by its respective parallax
- Use the information from 2a to determine appropriate colour cuts
- The final figure should have 
    - sensible axis limits
    - axis labels 
    - a labelled colour bar
    - a title
- Bonus: the size of each point should be inversely related to the measurement error of the parallax

````python
data = np.loadtxt('yaletrigplx.dat')
````

````python
# let's have a look at the shape of the data array
print(np.shape(data))
````

````python
# The five columns contain 
ID = data[:,0]
V = data[:,1]
color = data[:,2]
parallax = data[:,3]
parallax_err = data[:,4]
````

````python
plt.style.use('fast')
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.invert_yaxis()
sc = ax1.scatter(color, V, s=100./parallax_err, 
                 c=parallax, cmap=plt.cm.magma_r, vmin=0., vmax=0.25)
ax1.set_xlim(-0.5, 2.2)
ax1.set_xlabel('$B - V$')
ax1.set_ylabel('$V$')
cb = fig.colorbar(sc)
cb.set_label('Parallax [arcsec]')

fig.savefig('HR.png')
````
